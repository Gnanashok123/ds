```python
from sklearn.preprocessing import LabelEncoder
b=LabelEncoder()
c=["brand","model","link","type","engine_type"]
for values in c:
    a[values]=b.fit_transform(a[values])
a.info()
```

```python
a.isnull().sum()
```

```python
a['condition']=a['condition'].fillna(a['condition'].median())
a['speed_levels']=a['speed_levels'].fillna(a['speed_levels'].median())
a.isnull().sum()
```

```python
#Feature Engineering

a=a.drop('id',axis=1)
```

```python
#Data Visualization

#Box plot

import matplotlib.pyplot as plt
import seaborn as sn
sn.boxplot(x=a["highway_mileage"])
plt.show()
```

```python
#To check IQR(Inter Quartile Range)

Q1 = a["highway_mileage"].quantile(0.25)
Q3 = a["highway_mileage"].quantile(0.75)
IQR = Q3 - Q1
print(IQR)


#IQR <= 30 so need not to be change
```

```python
x=a.drop("brand",axis=1)
y=a["highway_mileage"]
x
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lnr=LinearRegression()
lnr.fit(x_train,y_train)
y_pred=lnr.predict(x_test)
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
